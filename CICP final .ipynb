### As df contain string data but model requires numeric data thats why we need to do label encoging and convert all the categorical data into numeric data for that we have to download labelencoder from sklearn.preprocessing 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df["workclass"]=le.fit_transform(df["workclass"])
df["marital.status"]=le.fit_transform(df["marital.status"])
df["occupation"]=le.fit_transform(df["occupation"])
df["relationship"]=le.fit_transform(df["relationship"])
df["race"]=le.fit_transform(df["race"])
df["sex"]=le.fit_transform(df["sex"])
df["native.country"]=le.fit_transform(df["native.country"])
df["income"]=le.fit_transform(df["income"])

df.head()
```

```python
x=df.drop(["income"], axis=1)
```

```python
y=df["income"]
```

```python
# Train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
gb.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# printing Metrics
## printing actual test and predicted test
```

```python
y_pred=gb.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred)*100)
```
